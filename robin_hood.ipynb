```python
# Bibliotecas utilizadas 

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

import pandas as pd
import time
import re 
```

# Busca Genérica 

Resulta um modelo de cada linha de celular. Exemplo: Motorola Edge Pro s

```python
columns = ['modelo','marca', 'lancamento', 'memoria_ram','memoria_interna',
           'tamanho_tela', 'resolucao_tela', 'resolucao_tela_direita', 'resolucao_tela_esquerda', 
           'resolucao_camera_principal','resolucao_camera_principal_direita','resolucao_camera_principal_esquerda',
           'resolucao_camera_frontal','resolucao_camera_frontal_direita','resolucao_camera_frontal_esquerda',
           'capacidade_bateria'] 

df = pd.DataFrame(columns=columns) # Criação do data frame, onde irá armazenar todos os dados

# Entra no site do adblock para ligarmos o bloqueador de anuncio. 
# Quando estiver no site do adblock, precisamos fazer a instalação clicando no 'Usar no Chrome', 
# apos isso aceitar o pop-up 'Adicionar extensão' e em 30s automaticamente entraremos no site para busca de informações
 
driver = webdriver.Chrome(executable_path='./chromedriver.exe')

driver.get("https://chrome.google.com/webstore/detail/adblock-%E2%80%94-best-ad-blocker/gighmmpiobklfepjocnamgkkbiglidom?hl=pt-BR")
                     
time.sleep(30)  

driver.get("https://www.maiscelular.com.br/fichas-tecnicas/?aparelho=1&z=0")
cookies = driver.find_element_by_xpath('//*[@id="frmMain"]/div[3]/div/div/a')
cookies.click()
time.sleep(3)
escopo = driver.find_element_by_xpath('/html/body/form/main/div[2]/div/div[2]/div[4]/span') #escopo => qtd de pg e pg atual
escopo = escopo.text
escopo = escopo.split()
pg_atual = escopo[5]
qtd_pg_total = int(escopo[7])
qtd_cell = int(escopo[1])
cont = 0

while (qtd_pg_total >= cont):
    for seq in range(1, qtd_cell+1):
        
        try:      
            modelo = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '//*[@id="accessibility_content"]/div[2]/div/div[2]/div[2]/div[{}]/div/div[1]/a/strong'.format(seq))))
            modelo = modelo.text                  

            nav = driver.find_element_by_xpath('//*[@id="accessibility_content"]/div[2]/div/div[2]/div[2]/div[{}]/div/div[1]/a/strong'.format(seq))
            nav.click()
        except: 
            try:
                spam = driver.find_element_by_xpath('//*[@id="dismiss-button"]/div/svg/path[1]')
                spam.click()
                time.sleep(3) 
            except:
                driver.back()


            modelo = driver.find_element_by_xpath('//*[@id="accessibility_content"]/div[2]/div/div[2]/div[2]/div[{}]/div/div[1]/a/strong'.format(seq))
            modelo = modelo.text

            nav = driver.find_element_by_xpath('//*[@id="accessibility_content"]/div[2]/div/div[2]/div[2]/div[{}]/div/div[1]/a/strong'.format(seq))
            nav.click()

        time.sleep(3)
        
        try:
            marca = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH,'/html/body/form/main/div[2]/div/div[1]/div[4]/div[3]/div[1]/table/tbody/tr[5]/td[2]/a')))
            marca = marca.text 
        except:
            marca = "brand"

        try:
            lancamento = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH,'/html/body/form/main/div[2]/div/div[1]/div[4]/div[3]/div[1]/table/tbody/tr[6]/td[2]')))
            lancamento = lancamento.text
        except:
            lancamento = "Não Informado"

        try:
            vendido = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH,'/html/body/form/main/div[2]/div/div[1]/div[4]/div[3]/div[1]/table/tbody/tr[4]/td[2]/small')))
            vendido = vendido.text 

        except:
            vendido = "Não Informado"

        try:
            memoria_ram = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH,'/html/body/form/main/div[2]/div/div[1]/div[4]/div[3]/div[1]/table/tbody/tr[21]/td[2]')))
            memoria_ram = memoria_ram.text 
        except:
            memoria_ram = "Valor não informado"
            memoria_ram = "Valor não informado"

        try:
            memoria_interna = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH,'/html/body/form/main/div[2]/div/div[1]/div[4]/div[3]/div[1]/table/tbody/tr[22]/td[2]')))
            memoria_interna = memoria_interna.text
        except:
            memoria_interna = "Valor não informado"
            memoria_interna = "Valor não informado"

        try:
            tamanho_tela = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH,'/html/body/form/main/div[2]/div/div[1]/div[4]/div[3]/div[1]/table/tbody/tr[26]/td[2]')))
            tamanho_tela = tamanho_tela.text
        except:
            tamanho_tela = "Valor não informado"
            tamanho_tela = "Valor não informado"

        try:
            resolucao_tela = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH,'/html/body/form/main/div[2]/div/div[1]/div[4]/div[3]/div[1]/table/tbody/tr[28]/td[2]')))
            resolucao_tela = resolucao_tela.text
            resolucao_tela_direita = resolucao_tela.split(" ")[0].replace("x", " ").split(" ")[1]
            resolucao_tela_esquerda = resolucao_tela.split(" ")[0].replace("x", " ").split(" ")[0]
        except:
            resolucao_tela = "Valor não informado"
            resolucao_tela = "Valor não informado"
            resolucao_tela_direita = "Valor não informado"
            resolucao_tela_esquerda = "Valor não informado"

        try:
            resolucao_camera_principal = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,'/html/body/form/main/div[2]/div/div[1]/div[4]/div[3]/div[1]/table/tbody/tr[57]/td[2]')))
            resolucao_camera_principal = resolucao_camera_principal.text
            resolucao_camera_principal_direita = resolucao_camera_principal.split(" ")[0].replace("x"," ").split(" ")[1]
            resolucao_camera_principal_esquerda  = resolucao_camera_principal.split(" ")[0].replace("x"," ").split(" ")[0]
        except:
            resolucao_camera_principal = "Valor não informado"
            resolucao_camera_principal = "Valor não informado"
            resolucao_camera_principal_direita = "Valor não informado"
            resolucao_camera_principal_esquerda = "Valor não informado"



        try:
            resolucao_camera_frontal = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH,'/html/body/form/main/div[2]/div/div[1]/div[4]/div[3]/div[1]/table/tbody/tr[73]/td[2]')))
            resolucao_camera_frontal = resolucao_camera_frontal.text
            resolucao_camera_frontal_direita = resolucao_camera_frontal.split(" ")[0].replace("x"," ").split(" ")[1]
            resolucao_camera_frontal_esquerda = resolucao_camera_frontal.split(" ")[0].replace("x"," ").split(" ")[0]
        except:
            resolucao_camera_frontal = "Valor não informado"
            resolucao_camera_frontal = "Valor não informado"
            resolucao_camera_frontal_direita = 0
            resolucao_camera_frontal_esquerda = 0    

        try:
            capacidade_bateria = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH,'/html/body/form/main/div[2]/div/div[1]/div[4]/div[3]/div[1]/table/tbody/tr[104]/td[2]')))
            capacidade_bateria = capacidade_bateria.text
        except:
            capacidade_bateria = "Valor não informado"
            capacidade_bateria = "Valor não informado"

            
        driver.back()
        time.sleep(3)

        df = df.append({'modelo':modelo,'marca':marca, 'lancamento':lancamento, 'memoria_ram':memoria_ram,
                        'memoria_interna':memoria_interna,'tamanho_tela':tamanho_tela, 'resolucao_tela':resolucao_tela, 
                        'resolucao_tela_direita':resolucao_tela_direita, 'resolucao_tela_esquerda':resolucao_tela_esquerda, 
                        'resolucao_camera_principal':resolucao_camera_principal,'resolucao_camera_principal_direita':resolucao_camera_principal_direita,
                        'resolucao_camera_principal_esquerda':resolucao_camera_principal_esquerda,
                        'resolucao_camera_frontal':resolucao_camera_frontal,'resolucao_camera_frontal_direita':resolucao_camera_frontal_direita,
                        'resolucao_camera_frontal_esquerda':resolucao_camera_frontal_esquerda,
                        'capacidade_bateria':capacidade_bateria}, ignore_index=True)


    cont+=1
    if pg_atual != qtd_pg_total:
        proxima = driver.find_element_by_xpath('/html/body/form/main/div[2]/div/div[2]/div[5]/a/span')
        proxima.click()
        escopo = driver.find_element_by_xpath('/html/body/form/main/div[2]/div/div[2]/div[4]/span') #escopo => qtd de pg e pg atual
        escopo = escopo.text
        escopo = escopo.split()
        qtd_cell = int(escopo[1])
        pg_atual = int(escopo[5])
    else:
        break 
else:
    print('finish')
    
df.to_excel('df.xlsx') # Exportar em .xlsx
```
